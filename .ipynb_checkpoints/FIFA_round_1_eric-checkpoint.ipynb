```python
import pandas as pd
pd.set_option('display.max_columns', None)
data = pd.read_csv('fifa21_train.csv')
```

```python
data.head().T
```

```python
data.drop_duplicates()
```

```python
# Probably useless columns in the modelisation (redondant info or unique)
data = data.drop(['ID', 'Name', 'Team & Contract'], axis=1)
```

```python
# Let's apply lower case to all remaining column names and replace spaces by underscore
data.columns = [colname.lower().replace(' ','_') for colname in data.columns]
data.columns
```

```python
# code found here :  https://stackoverflow.com/questions/37333299/splitting-a-pandas-dataframe-column-by-delimiter
for col in ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf',
       'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm',
       'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk']:
    extra_col = col + '_evol'
    data[[col,extra_col]] = data[col].str.split('+',expand=True).astype(float)
```

```python
data.dtypes
```

```python
# remove useless letters in numerical columns (like lbs in weight)
data['weight'] = pd.to_numeric(data['weight'].map(lambda x: x.rstrip('lbs')))
```

```python
data['value'] = data['value'].str.lstrip("€")
```

```python
data['value'] = data['value'].apply(lambda x: x.rstrip('K') if x.strip()[-1] == "K" else float(x.rstrip('M')) * 1000)
```

```python
data['value'].head()
```

```python
# Let's use numerical labels on the 'A/W' and 'D/W'
work_rate_mapper = { 'Low':1, 'Medium':2, 'High':3 }
data['a/w'] = data['a/w'].replace(work_rate_mapper)
data['d/w'] = data['d/w'].replace(work_rate_mapper)
data['a/w'].head()
```

```python
for c in ['ir', 'w/f', 'sm']:
    data[c] = data[c].str[:1].astype(int)
data['w/f'].head()
```
